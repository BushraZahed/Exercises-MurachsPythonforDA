# Adding some visualization
## 3. Use the wide data to create a line plot for just the age group 15-19 Years , include an appropriate title and remove the legend

```python
mortality_wide.plot.line(title='Child Mortality: AgeGroup of 15-19 Years ', y = ['15-19 Years'],
                         ylabel='Deaths per 100,000', legend = False, grid=True, rot=45)
```

## 4. Use the wide data to create an area plot for all age groups, and reverse the order of the items in the legend to see how that looks

```python
mortality_wide.plot.area(title = "Area plot of Child Mortality: for all AgeGroup from 01-19 Years", ylabel='Deaths per 100,000', legend = 'reverse',rot = 45)
```

## 5. Use the wide data to create a bar plot for all age groups that shows the mortality rates for just the year 1900, and note the value on the y- axis.

```python
mortality_wide.query("Year == 1900").plot.bar(title = "Bar plot of Child Mortality: for all AgeGroup from 01-19 Years", ylabel='Deaths per 100,000', legend = True, rot = 45)
```

## 6. Change the bar plot you created in step 5 to show the mortality rates for the year 2000 and note how the value of the y-axis change.Then add an appropriate title to the plot and remove the label for the x axis.

```python
mortality_wide.query("Year == 2000").plot.bar(title = "Bar plot of Child Mortality: for all AgeGroup from 01-19 Years for the year 2000", xlabel= "", ylabel='Deaths per 100,000',legend = True, rot = 45)
```

## 7. Use the long data to create a bar plot like in step 6. To do that, you will need to chain the pivot method to the query method. Compare the two bar charts,and then make improvements so the plot that uses long data is easier to read.

```python
mortality_data.query("Year == 2000").pivot(index = "AgeGroup", columns = "Year", values = "DeathRate").plot.bar(title = "Bar plot of Child Mortality: for all AgeGroup from 01-19 Years for the Year 2000", xlabel= "", ylabel='Deaths per 100,000',legend = True, rot = 45)
```

## 9. Use the long data to create a density plot that shows the distribution of the death rates in the year 1900. Include a title and grid in the plot to make the data easier to read.

```python
mortality_data.query("Year == 1900").plot.density(y = 'DeathRate', grid = True, title = "density plot that shows the distribution of the death rates in the year 1900")
```

## 10. Use  the long data to create a pie plot, that shows the sum of the death rates for the years 1900, 1925, 1950, 1975, 2000

```python
mortality_data.query(' Year in (1900, 1925, 1950, 1975, 2000)').groupby('Year')['DeathRate'].sum().plot.pie()
```
